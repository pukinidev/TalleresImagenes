## Utilizando filtros lineales o no, enmarque la figura de un contorno claro (blanco o casi blanco) (conservando lo mejor posible los niveles de gris originales de la imagen).

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

from skimage import exposure
from skimage.filters import gaussian
from skimage.filters import sobel_h, sobel_v, sobel
from skimage.filters import laplace
```

```python
amour = cv2.imread("images/amour.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(amour, cmap= 'gray', vmin = 0, vmax= 255)
plt.show()
```

```python
thres_value, amour_s = cv2.threshold(amour,  amour.mean(), 255, cv2.THRESH_BINARY)

plt.imshow(amour_s, cmap= 'gray')

plt.show()
```

```python
amour_s = sobel(amour)
amour_s = exposure.rescale_intensity(amour_s, in_range=(0, 255))

thres_value, amour_s = cv2.threshold(amour_s,  0.00025, 255, cv2.THRESH_BINARY)

plt.imshow(amour_s, cmap= 'gray')

plt.show()
```

```python
amour_s.max()
```

```python
amour_s = amour_s.astype(np.uint8)
amour = amour.astype(np.uint8)


final = cv2.bitwise_or(amour, amour_s)

plt.imshow(final, cmap= 'gray',  vmin = 0, vmax= 255)
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (20,15)) 
axes[0].imshow(amour, cmap= 'gray', vmin = 0, vmax= 255)
axes[0].set_title('Original')
axes[1].imshow(final, cmap= 'gray', vmin = 0, vmax= 255)
axes[1].set_title('Resultado')
plt.show()
```
