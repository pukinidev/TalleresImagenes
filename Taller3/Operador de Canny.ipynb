```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import sobel
from skimage.feature import canny
```

# Cargar Imagen

```python
rondelle = cv2.imread('images/rondelle.png', cv2.IMREAD_GRAYSCALE)
```

```python
sobel = sobel(rondelle)
plt.imshow(sobel, cmap='gray')
```

# Operador Canny

```python
edges = canny(rondelle,low_threshold=0, high_threshold=1,  sigma=10)
plt.imshow(edges, cmap='gray')
```

## Explicación

Al utilizar una desviación estándar de 10, se puede observar que la detección de bordes resulta efectiva, ya que permite suavizar el ruido presente en la imagen y resaltar los contornos principales del objeto. En este caso, los bordes de la rondela y su agujero central se detectan claramente, mostrando una forma bien definida. Sin embargo, se alcanza a notar ruido dentro de la rondela y en algunas regiones alrededor del anillo, esto puede ser causado debido a que la imagen cuenta con un ruido aditivo considerable.

# Desviación Estándar en 5

```python
edges_5 = canny(rondelle,low_threshold=0, high_threshold=1,  sigma=5)
plt.imshow(edges_5, cmap='gray')
```

## Explicación

Al utilizar una desviación estándar de 5, se puede observar que la detección de bordes no resulta tan efectiva, ya que a pesar permite resaltar los contornos principales del objeto es posible identificar cantidades considerables de ruido.

# Umbral Superior en 5 

```python
umbral_5 = canny(rondelle,low_threshold=0, high_threshold=5,  sigma=10)
plt.imshow(umbral_5, cmap='gray')
```

```python
umbral_5 = canny(rondelle,low_threshold=0, high_threshold=5,  sigma=5)
plt.imshow(umbral_5, cmap='gray')
```

## Explicación

Utilizando un umbral superior de 5 y una desviación estándar de 10 y 5, se puede observar que la detección de contornos mejora notablemente. Esto se debe a que al aumentar el umbral, se eliminan la mayoría de las falsas detecciones causadas por pequeñas variaciones en la amplitud del gradiente. Por ejemplo, con la imagen que cuenta con una una desviación estándar de 10 se puede observar los contornos de la imagen con gran claridad y con una cantidad no significativa de ruido dentro de la rondela. Mientras que con la imagen que cuenta con una desviación estándar de 5, se observan más detalles en los bordes, pero también puede haber una ligera presencia de ruido dentro de la rondela.
