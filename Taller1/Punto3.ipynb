# Negativo

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('/home/valiwis/Documents/Maestria/Imagenes y Vision/Taller 1/images/image_punto3.png')
```

```python
plt.imshow(image)
```

```python
hist = cv2.calcHist([image], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```

```python
not_image = cv2.bitwise_not(image)
```

```python
plt.imshow(not_image)
```

```python
hist = cv2.calcHist([not_image], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```

## Expansion


```python
min_val = np.min(image)
max_val = np.max(image)

imagen_contraste_expandido = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)
plt.imshow(imagen_contraste_expandido)
```

```python
hist = cv2.calcHist([imagen_contraste_expandido ], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()
```
