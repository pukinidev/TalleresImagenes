# An√°lisis del histograma

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import urllib.request
```

```python
def process_image(image_path):
    req = urllib.request.urlopen(image_path)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv2.imdecode(arr, -1)
    return img
```

```python
clef_image = process_image('https://raw.githubusercontent.com/pukinidev/TalleresImagenes/refs/heads/main/Taller1/images/clef.png')
muscle_image = process_image('https://raw.githubusercontent.com/pukinidev/TalleresImagenes/refs/heads/main/Taller1/images/muscle.png')
```

```python
plt.imshow(clef_image, cmap='gray')
```

```python
plt.imshow(muscle_image, cmap='gray')
```

## Histograma

```python
hist = cv2.calcHist([clef_image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
hist = cv2.calcHist([muscle_image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

## Explicaci√≥n

Para el histograma numero 1, encontramos que el nivel de gris se encuentra en un rango de 100-150 por lo que podriamos decir que se encuentra en un rango de tonos medios.

En el segundo histograma, la distribuci√≥n de los tonos de gris es mayor en las zonas oscuras, con una notable presencia en los tonos medios y una cantidad m√≠nima en los tonos m√°s claros.

C√≥mo se puede determinar cu√°les fibras musculares son m√°s densas (ocupan una superficie m√°s grande en la imagen) entre las rojas (oscuras) y las blancas (grises)? Y ¬øcu√°les fibras tienen un color m√°s uniforme (m√°s homog√©neo)?

Para determinar las fibras que son mas densas se podrian observar los picos del histograma que indican cuantos pixeles tienen un determinado valor de intensidad.

En este caso:
- Primer pico (bajo 50): Representa las fibras oscuras (rojas), ya que tienen valores bajos de intensidad.
- Segundo pico (entre 150 y 200): Representa las fibras claras (grises).

Para determinar las que tienen un color mas uniforme se podria observar el ancho de los picos ya que esto indica que la mayor√≠a de los p√≠xeles tienen intensidades similares. 

En este caso las fibras que tienen un color mas uniforme corresponde a las fibras oscura ya que el pico es pronunciado.
