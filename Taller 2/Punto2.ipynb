# Ecualizaci√≥n del histograma

```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagen_quito = cv2.imread('images/QUITO.png', cv2.IMREAD_GRAYSCALE)
imagen_baboon = cv2.imread('images/BABOON.png', cv2.IMREAD_GRAYSCALE)
```

```python
def mostrar_histograma(imagen):
    hist = cv2.calcHist([imagen],[0],None,[256],[0,256])
    plt.plot(hist)
    plt.show()
```

```python
hist_quito = mostrar_histograma(imagen_quito)
```

```python
plt.imshow(imagen_quito, cmap='gray', vmin=0, vmax=255)
```

## Ecualizaci√≥n de la imagen QUITO

```python
image_quito_ecu = cv2.equalizeHist(imagen_quito)
```

```python
plt.imshow(image_quito_ecu, cmap='gray', vmin=0, vmax=255)
```

```python
mostrar_histograma(image_quito_ecu)
```

## Segunda equalizaci√≥n de la imagen QUITO

```python
segundo_quito_ecu = cv2.equalizeHist(image_quito_ecu)
```

```python
plt.imshow(segundo_quito_ecu, cmap='gray', vmin=0, vmax=255)
```

```python
mostrar_histograma(segundo_quito_ecu)
```
